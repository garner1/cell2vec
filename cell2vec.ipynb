```python
import numpy as np
from scipy import ndimage
import imageio
import matplotlib.pyplot as plt
import os

img = imageio.imread('./Classification_single_cells copy/Cancer/Single_cells_dapiimage_186_cell_no_614.jpg')
# plt.imshow(img[rmin:rmax,cmin:cmax])      
# plt.show()
```

```python
def bbox(img):
    rows = np.any(img, axis=1)
    cols = np.any(img, axis=0)
    rmin, rmax = np.where(rows)[0][[0, -1]]
    cmin, cmax = np.where(cols)[0][[0, -1]]

    return rmin, rmax, cmin, cmax
bbox(img)
```

```python
directory = './Classification_single_cells copy/Cancer'
for img in os.listdir('./Classification_single_cells copy/Cancer')[:10]:
    filename = os.path.join(directory, img)
    img = imageio.imread(filename)
    rmin, rmax, cmin, cmax = bbox(img)
#     print(np.size(img[rmin:rmax,cmin:cmax]))
    plt.imshow(img[rmin:rmax,cmin:cmax])
    plt.show()
```
